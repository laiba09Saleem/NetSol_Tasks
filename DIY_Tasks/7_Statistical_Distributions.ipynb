# Statistical Distributions Visualization
This notebook demonstrates 7 important statistical distributions with visualizations. Each distribution is explained and plotted for better understanding


```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Set style for better looking plots
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")
```

**1. Uniform Distribution**

**What is it?**
A uniform distribution is where all outcomes are equally likely within a specified range. Think of rolling a fair die - each number from 1 to 6 has the same chance of appearing.

**Real-world example:** Random number generation, rolling dice, lottery numbers

```python
# Generate uniform distribution
low = 0  # Minimum value
high = 10  # Maximum value
size = 1000  # Number of samples

uniform_data = np.random.uniform(low, high, size)

# Create visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Histogram
ax1.hist(uniform_data, bins=20, edgecolor='black', alpha=0.7, color='blue')
ax1.set_title('Uniform Distribution - Histogram')
ax1.set_xlabel('Value')
ax1.set_ylabel('Frequency')
ax1.axvline(x=low, color='red', linestyle='--', label='Lower bound')
ax1.axvline(x=high, color='green', linestyle='--', label='Upper bound')
ax1.legend()

# Probability density function
x = np.linspace(low - 1, high + 1, 100)
y = stats.uniform.pdf(x, loc=low, scale=high-low)
ax2.plot(x, y, 'b-', linewidth=2, label='PDF')
ax2.fill_between(x, 0, y, alpha=0.3, color='blue')
ax2.set_title('Uniform Distribution - PDF')
ax2.set_xlabel('Value')
ax2.set_ylabel('Probability Density')
ax2.legend()


plt.tight_layout()
plt.show()

print(f"Uniform Distribution Summary:")
print(f"Range: {low} to {high}")
print(f"Mean: {np.mean(uniform_data):.2f}")
print(f"Standard Deviation: {np.std(uniform_data):.2f}")
print(f"Min value: {np.min(uniform_data):.2f}")
print(f"Max value: {np.max(uniform_data):.2f}")
```

**2. Binomial Distribution**

**What is it?**
Models the number of successes in a fixed number of trials, each with the same probability of success.

**Real-world example:** Flipping a coin 10 times and counting heads, quality control inspection


```python
# Parameters
n_trials = 20  # Number of trials (e.g., 20 coin flips)
p_success = 0.5  # Probability of success (e.g., probability of heads)
size = 1000  # Number of experiments

# Generate binomial distribution
binomial_data = np.random.binomial(n_trials, p_success, size)

# Create visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Histogram
ax1.hist(binomial_data, bins=range(n_trials+2), edgecolor='black', alpha=0.7, color='green')
ax1.set_title(f'Binomial Distribution - {n_trials} Trials, p={p_success}')
ax1.set_xlabel('Number of Successes')
ax1.set_ylabel('Frequency')

# Probability mass function
x = np.arange(0, n_trials+1)
y = stats.binom.pmf(x, n_trials, p_success)
ax2.bar(x, y, alpha=0.7, color='green', edgecolor='black')
ax2.set_title('Binomial Distribution - PMF')
ax2.set_xlabel('Number of Successes')
ax2.set_ylabel('Probability')

plt.tight_layout()
plt.show()

print(f"Binomial Distribution Summary:")
print(f"Number of trials: {n_trials}")
print(f"Probability of success: {p_success}")
print(f"Expected successes: {n_trials * p_success}")
print(f"Mean successes: {np.mean(binomial_data):.2f}")
print(f"Standard deviation: {np.std(binomial_data):.2f}")
```

**3. Bernoulli Distribution**

**What is it?**
Models a single trial with only two possible outcomes: success (1) or failure (0).
This is a special case of the binomial distribution where n=1.

**Real-world example:** A single coin flip, whether a customer buys a product (yes/no)

```python
# Parameters
p_success = 0.3  # Probability of success
size = 1000  # Number of trials

# Generate Bernoulli distribution
bernoulli_data = np.random.binomial(1, p_success, size)

# Create visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Bar plot of outcomes
outcomes = ['Failure (0)', 'Success (1)']
counts = [np.sum(bernoulli_data == 0), np.sum(bernoulli_data == 1)]
ax1.bar(outcomes, counts, color=['red', 'green'], alpha=0.7, edgecolor='black')
ax1.set_title(f'Bernoulli Distribution - p={p_success}')
ax1.set_ylabel('Frequency')

# Probability mass function
x = [0, 1]
y = [1-p_success, p_success]
ax2.bar(x, y, alpha=0.7, color=['red', 'green'], edgecolor='black')
ax2.set_title('Bernoulli Distribution - PMF')
ax2.set_xlabel('Outcome')
ax2.set_ylabel('Probability')
ax2.set_xticks(x)
ax2.set_xticklabels(['Failure (0)', 'Success (1)'])

plt.tight_layout()
plt.show()

print(f"Bernoulli Distribution Summary:")
print(f"Probability of success: {p_success}")
print(f"Probability of failure: {1-p_success}")
print(f"Number of successes: {np.sum(bernoulli_data)}")
print(f"Number of failures: {np.sum(bernoulli_data == 0)}")
print(f"Observed success rate: {np.mean(bernoulli_data):.3f}")
```

**4. Poisson Distribution**

**What is it?**
Models the count of events in a fixed interval, given a constant average rate.

**Real-world example:** Number of emails received per hour, customers arriving at a store

```python
# Parameters
lambda_rate = 3  # Average number of events per interval
size = 1000  # Number of samples

# Generate Poisson distribution
poisson_data = np.random.poisson(lambda_rate, size)

# Create visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Histogram
ax1.hist(poisson_data, bins=range(0, max(poisson_data)+2), edgecolor='black', alpha=0.7, color='orange')
ax1.set_title(f'Poisson Distribution - λ={lambda_rate}')
ax1.set_xlabel('Number of Events')
ax1.set_ylabel('Frequency')

# Probability mass function
x = np.arange(0, 15)
y = stats.poisson.pmf(x, lambda_rate)
ax2.bar(x, y, alpha=0.7, color='orange', edgecolor='black')
ax2.set_title('Poisson Distribution - PMF')
ax2.set_xlabel('Number of Events')
ax2.set_ylabel('Probability')

plt.tight_layout()
plt.show()

print(f"Poisson Distribution Summary:")
print(f"Lambda (average rate): {lambda_rate}")
print(f"Mean events: {np.mean(poisson_data):.2f}")
print(f"Variance: {np.var(poisson_data):.2f}")
print(f"Standard deviation: {np.std(poisson_data):.2f}")
```

**5. Normal Distribution**

**What is it?**
A symmetric, bell-shaped distribution where most values cluster around the mean.

**Real-world example:** Height of people, test scores, measurement errors

```python
# Parameters
mean = 0  # Mean of the distribution
std_dev = 1  # Standard deviation
size = 1000  # Number of samples

# Generate normal distribution
normal_data = np.random.normal(mean, std_dev, size)

# Create visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Histogram with density curve
ax1.hist(normal_data, bins=30, density=True, edgecolor='black', alpha=0.7, color='purple')
x = np.linspace(mean - 3*std_dev, mean + 3*std_dev, 100)
y = stats.norm.pdf(x, mean, std_dev)
ax1.plot(x, y, 'r-', linewidth=2, label='Normal Curve')
ax1.set_title(f'Normal Distribution - μ={mean}, σ={std_dev}')
ax1.set_xlabel('Value')
ax1.set_ylabel('Density')
ax1.legend()

# Q-Q plot
stats.probplot(normal_data, dist="norm", plot=ax2)
ax2.set_title('Q-Q Plot - Check for Normality')

plt.tight_layout()
plt.show()

print(f"Normal Distribution Summary:")
print(f"Mean: {np.mean(normal_data):.2f}")
print(f"Standard deviation: {np.std(normal_data):.2f}")
print(f"Skewness: {stats.skew(normal_data):.3f} (should be close to 0)")
print(f"Kurtosis: {stats.kurtosis(normal_data):.3f} (should be close to 0)")
print(f"68% of data falls within ±1 std dev: {np.mean(np.abs(normal_data) <= 1)*100:.1f}%")
print(f"95% of data falls within ±2 std dev: {np.mean(np.abs(normal_data) <= 2)*100:.1f}%")
```

**6. T-Distribution**

**What is it?**
Used in t-tests to compare sample means, especially with small sample sizes or unknown population variance.
It's similar to normal distribution but has heavier tails.

**Real-world example:** Comparing means of two small samples, confidence intervals

```python
# Parameters
df = 5  # Degrees of freedom (n-1)
size = 1000  # Number of samples

# Generate t-distribution
t_data = np.random.standard_t(df, size)

# Create visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Histogram with density curve
ax1.hist(t_data, bins=30, density=True, edgecolor='black', alpha=0.7, color='red')
x = np.linspace(-4, 4, 100)
y_t = stats.t.pdf(x, df)
y_normal = stats.norm.pdf(x, 0, 1)
ax1.plot(x, y_t, 'r-', linewidth=2, label=f't-distribution (df={df})')
ax1.plot(x, y_normal, 'b--', linewidth=2, label='Normal Distribution')
ax1.set_title('t-Distribution vs Normal Distribution')
ax1.set_xlabel('Value')
ax1.set_ylabel('Density')
ax1.legend()

# Q-Q plot
stats.probplot(t_data, dist="t", sparams=(df,), plot=ax2)
ax2.set_title(f't-Distribution Q-Q Plot (df={df})')

plt.tight_layout()
plt.show()

print(f"t-Distribution Summary:")
print(f"Degrees of freedom: {df}")
print(f"Mean: {np.mean(t_data):.2f}")
print(f"Standard deviation: {np.std(t_data):.2f}")
print(f"Comparison with normal distribution:")
print(f"- t-distribution has heavier tails")
print(f"- As df increases, t-distribution approaches normal distribution")
```

**7. Exponential Distribution**

**What is it?**
Models the time between events in a process with a constant average rate.

**Real-world example:** Time between customer arrivals, lifetime of electronic components

```python
# Parameters
rate = 0.5  # Rate parameter (1/mean)
scale = 1/rate  # Scale parameter
size = 1000  # Number of samples

# Generate exponential distribution
exp_data = np.random.exponential(scale, size)

# Create visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Histogram
ax1.hist(exp_data, bins=30, edgecolor='black', alpha=0.7, color='brown')
ax1.set_title(f'Exponential Distribution - λ={rate}')
ax1.set_xlabel('Time Between Events')
ax1.set_ylabel('Frequency')

# Probability density function
x = np.linspace(0, max(exp_data)*1.1, 100)
y = stats.expon.pdf(x, scale=scale)
ax2.plot(x, y, 'b-', linewidth=2, label=f'PDF (λ={rate})')
ax2.fill_between(x, 0, y, alpha=0.3, color='brown')
ax2.set_title('Exponential Distribution - PDF')
ax2.set_xlabel('Time')
ax2.set_ylabel('Probability Density')
ax2.legend()

plt.tight_layout()
plt.show()

print(f"Exponential Distribution Summary:")
print(f"Rate (λ): {rate}")
print(f"Mean: {np.mean(exp_data):.2f} (theoretical: {1/rate:.2f})")
print(f"Median: {np.median(exp_data):.2f} (theoretical: {np.log(2)/rate:.2f})")
print(f"Standard deviation: {np.std(exp_data):.2f}")
print(f"Memoryless property: P(X > s+t | X > s) = P(X > t)")
```

**Comparison of All 7 Distributions**

Let's visualize all distributions together to compare their shapes

```python
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.flatten()

# Uniform Distribution
uniform_data = np.random.uniform(0, 10, 1000)
axes[0].hist(uniform_data, bins=20, alpha=0.7, color='blue')
axes[0].set_title('Uniform Distribution')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Frequency')

# Binomial Distribution
binomial_data = np.random.binomial(20, 0.5, 1000)
axes[1].hist(binomial_data, bins=range(22), alpha=0.7, color='green')
axes[1].set_title('Binomial Distribution')
axes[1].set_xlabel('Successes')

# Bernoulli Distribution
bernoulli_data = np.random.binomial(1, 0.3, 1000)
axes[2].hist(bernoulli_data, bins=[-0.5, 0.5, 1.5], alpha=0.7, color='red')
axes[2].set_title('Bernoulli Distribution')
axes[2].set_xticks([0, 1])
axes[2].set_xlabel('Outcome')

# Poisson Distribution
poisson_data = np.random.poisson(3, 1000)
axes[3].hist(poisson_data, bins=range(12), alpha=0.7, color='orange')
axes[3].set_title('Poisson Distribution')
axes[3].set_xlabel('Events')

# Normal Distribution
normal_data = np.random.normal(0, 1, 1000)
axes[4].hist(normal_data, bins=30, alpha=0.7, color='purple')
axes[4].set_title('Normal Distribution')
axes[4].set_xlabel('Value')

# t-Distribution
t_data = np.random.standard_t(5, 1000)
axes[5].hist(t_data, bins=30, alpha=0.7, color='red')
axes[5].set_title('t-Distribution')
axes[5].set_xlabel('Value')

# Exponential Distribution
exp_data = np.random.exponential(2, 1000)
axes[6].hist(exp_data, bins=30, alpha=0.7, color='brown')
axes[6].set_title('Exponential Distribution')
axes[6].set_xlabel('Time')

# Remove empty subplot
axes[7].axis('off')
axes[8].axis('off')

plt.suptitle('Comparison of All 7 Statistical Distributions', fontsize=16)
plt.tight_layout()
plt.show()
```

**Interactive Distribution Explorer**

**Try changing the parameters!**
This interactive cell lets you explore how different parameters affect each distribution

```python


from ipywidgets import interact, interactive, fixed, widgets

def explore_distribution(dist_type, param1, param2, sample_size):
    if dist_type == 'Normal':
        data = np.random.normal(param1, param2, sample_size)
        title = f'Normal Distribution (μ={param1}, σ={param2})'
    elif dist_type == 'Binomial':
        n = int(param1)
        p = param2
        data = np.random.binomial(n, p, sample_size)
        title = f'Binomial Distribution (n={n}, p={p})'
    elif dist_type == 'Poisson':
        data = np.random.poisson(param1, sample_size)
        title = f'Poisson Distribution (λ={param1})'
    elif dist_type == 'Exponential':
        data = np.random.exponential(param1, sample_size)
        title = f'Exponential Distribution (scale={param1})'
    else:
        return
    
    plt.figure(figsize=(10, 6))
    plt.hist(data, bins=30, alpha=0.7, edgecolor='black')
    plt.title(title)
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.grid(True, alpha=0.3)
    plt.show()
    
    print(f"Sample Statistics:")
    print(f"Mean: {np.mean(data):.3f}")
    print(f"Median: {np.median(data):.3f}")
    print(f"Std Dev: {np.std(data):.3f}")

# Create interactive widget
interact(explore_distribution,
         dist_type=widgets.Dropdown(options=['Normal', 'Binomial', 'Poisson', 'Exponential'],
                                   description='Distribution:'),
         param1=widgets.FloatSlider(min=0, max=10, step=0.1, value=0, description='Param 1:'),
         param2=widgets.FloatSlider(min=0, max=5, step=0.1, value=1, description='Param 2:'),
         sample_size=widgets.IntSlider(min=100, max=10000, step=100, value=1000, description='Sample Size:'))
```
