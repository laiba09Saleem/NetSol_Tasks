```
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from ipywidgets import interact
np.random.seed(42)
```

## 📈 1. Visualizing Distributions

```
data = np.random.normal(0, 1, 1000)
plt.hist(data, bins=30)
plt.title("Normal Distribution")
plt.show()
```

### 🎛 Interactive Normal Distribution

```
def plot_normal(mu=0, sigma=1):
    x = np.linspace(-10, 10, 1000)
    y = (1/(sigma*np.sqrt(2*np.pi))) * np.exp(-0.5*((x-mu)/sigma)**2)
    plt.plot(x, y)
    plt.title(f"μ={mu}, σ={sigma}")
    plt.show()

interact(plot_normal, mu=(-5,5,1), sigma=(1,5,1))
```

## 🧪 2. Hypothesis Testing

```
group_A = np.random.normal(70, 10, 50)
group_B = np.random.normal(75, 10, 50)

t_stat, p_val = stats.ttest_ind(group_A, group_B)
print("T-test p-value:", p_val)
```

## 📊 3. Multiple Statistical Tests

```
# Normality
print("Shapiro:", stats.shapiro(group_A))

# Mann-Whitney
print("Mann-Whitney:", stats.mannwhitneyu(group_A, group_B))

# ANOVA
group_C = np.random.normal(80, 10, 50)
print("ANOVA:", stats.f_oneway(group_A, group_B, group_C))
```

## 📉 4. Correlation Analysis

```
hours = np.random.randint(1, 10, 50)
scores = hours * 5 + np.random.normal(0, 5, 50)

plt.scatter(hours, scores)
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.show()

print("Pearson:", stats.pearsonr(hours, scores))
```
