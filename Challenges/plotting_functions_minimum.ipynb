# Plotting Quadratic and Non-Quadratic Functions with Minimum Value

## Objective
The objective of this task is to:
- take one quadratic function
- take one non-quadratic function
- plot both functions against x
- find their minimum values using Python

* Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

**Quadratic Function Statement**

## Quadratic Function

A quadratic function has degree 2 and is generally written as:

y = ax^2 + bx + c


For this task, we use:
y = x^2 - 4x + 3

**Quadratic Function Code**

```python
# x values
x = np.linspace(-2, 6, 100)

# quadratic function
y_quad = x**2 - 4*x + 3

# minimum value
min_index_quad = np.argmin(y_quad)
min_x_quad = x[min_index_quad]
min_y_quad = y_quad[min_index_quad]

print("Quadratic Function")
print("Minimum x:", min_x_quad)
print("Minimum y:", min_y_quad)

# plot
plt.figure(figsize=(8,5))
plt.plot(x, y_quad, label='y = x^2 - 4x + 3')
plt.scatter(min_x_quad, min_y_quad, s=100, label='Minimum Point')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Quadratic Function")
plt.grid(True)
plt.legend()
plt.show()
```

**Result Statement**

### Observation
The graph shows a parabola opening upward.

The minimum value occurs at the vertex of the parabola.
